### Program to display Normal, Binomial, Poisson, Bernoulli distributions for a given frequency distribution and analyse the results.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, binom, poisson, bernoulli
```

```python
# Set style for plots 
sns.set(style =  "whitegrid")
```

```python
# 1. Normal Distribution
X = np.linspace(-2, 2, 15)
y = norm.pdf(X, loc = 0, scale = 1)
plt.figure(figsize = (12,8))
plt.subplot(2, 2, 1)
plt.plot(X, y, 'b', label = "Mean = 0, SD = 1")
plt.title("Normal Distribution")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.legend()
```

```python
# 2. Binomial Distribution
n, p = 10, 0.5
X = np.arange(0, n+1)
y = binom.pmf(X, n, p)
plt.subplot(2, 2, 2)
plt.stem(X, y, basefmt = " ")
plt.title("Binomial Distribution (n = 10, p = 0.5)")
plt.xlabel("Number of successes")
plt.ylabel("Probability")
```

```python
# 3. Poisson Distribution
lam = 4
X = np.arange(0, 15)
y = poisson.pmf(X, lam)
plt.subplot(2, 2, 3)
plt.stem(X, y, basefmt = " ")
plt.title("Poisson Distribution ( = 4)")
plt.xlabel("Number of events")
plt.ylabel("Probability")
```

```python
# 4. Bernoulli Distribution
p = 0.6
X = [0, 1]
y = bernoulli.pmf(X, p)
plt.subplot(2, 2, 4)
plt.stem(X, y, basefmt = " ")
plt.title("Bernoulli Distribution (p = 0.6)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.xticks([0, 1], ["Failure(0)", "Success(1)"])
```

Program to implement correlation, rank correlation and regression and plot x-y plot and heat maps of correlation matrices.

```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
from sklearn.linear_model import LinearRegression
```

```python
# 1. Create sample data
data = {
    'Study_Hours' : [2, 3, 4, 5, 6, 7, 8, 9],
    'Marks' : [45, 50, 55, 60, 65, 68, 72, 75],
    'Attendance' : [60, 65, 70, 75, 80, 85, 90, 95]
}

df = pd.DataFrame(data)
print("Dataset")
print(df)
```

```python
# 2. Pearson Correlation
corr_matrix = df.corr(method = 'pearson')
print("\nPearson Correlation Matrix")
print(corr_matrix)
```

```python
# 3. Spearman Rank Correlation
rho, pval = spearmanr(df['Study_Hours'], df['Marks'])
print("\nSpearman Rank Correlation (Study Hours vs Marks)")
print("Spearman's rho : ", rho)
print("p-value : ", pval)
```

```python
# 4. Linear Regression
x = df[['Study_Hours']]
y = df['Marks']
model = LinearRegression()
model.fit(x,y)

# Predict and Plot
df['Predicted_Marks'] = model.predict(x)
plt.scatter(df['Study_Hours'], df['Marks'], color = 'blue', label = "Actual")
plt.plot(df['Study_Hours'], df['Predicted_Marks'], color = 'red', label = 'Regression Line')
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5. Heatmap of Correlation
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm')
plt.title('Heatmap of correlation matrix')
plt.show()
```

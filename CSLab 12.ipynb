Program to implement multiple linear regression using iris dataset, visualize and analyze the results.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("<your_path>iris.csv")
print("First 5 rows of uploaded data")
print(df.head())
```

```python
X = df[['sepal_length', 'sepal_width', 'petal_length']]
y = df['petal_width']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size =0.2, random_state = 1
)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("\nModel Coefficients")
print("Intercept: ", model.intercept_)
print("Coefficients: ", model.coef_)
print("\nEvaluation Metrics")
print("MeanSquaredError: ", mean_squared_error(y_test, y_pred))
print("R2 Score: ", r2_score(y_test, y_pred))
```

```python
plt.scatter(y_test, y_pred, color = 'green')
plt.plot([0, 2.5], [0, 2.5], 'r--')
plt.xlabel("Actual Petal Width")
plt.ylabel("Predicted Petal Width")
plt.title("Actual vs Predicted (Multiple Linear Regression)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Program to implement the working of linear discriminant analysis using iris dataset and visualize the results.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
data = pd.read_csv("<your_path>iris.csv")
print("First 5 rows of uploaded data")
print(data.head())
```

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

print("Shape of features (X): ", X.shape)
print("Shape of labels (y): ", y.shape)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
lda = LDA(n_components = 2)
X_lda = lda.fit_transform(X_scaled, y)
print("Shape after LDA")
print(X_lda.shape)

plt.figure(figsize = (8, 6))
for label, color in zip(set(y), ['red', 'green', 'blue']):
    plt.scatter(
        X_lda[y == label, 0],
        X_lda[y == label, 1],
        alpha = 0.7,
        label = label,
        color = color
    )

plt.xlabel('LDA Component 1')
plt.ylabel('LDA component 2')
plt.title('LDA: iris dataset (Manual CSV upload)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
